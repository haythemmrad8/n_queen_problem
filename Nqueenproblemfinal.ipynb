# Placer reine :

```python
def placer_reine(cord,n,P):
    x,y=cord
    L={(k,y)for k in range (0,n)}
    if set(P).intersection(L): return False
    M={(x,k)for k in range (0,n)}
    if set(P).intersection(M): return False
    N={(x+k,y-k) for k in range(-x,n-x)}
    if set(P).intersection(N): return False
    O={(x-k,y-k) for k in range(x,x-n,-1)}
    if set(P).intersection(O): return False
    return True
```

# placer toutes reines :

```python
def placer_tous_reines(n):
    x,y,L,P=0,0,[],[]
    while True:
        while x<n:
            if placer_reine((x,y),n,P):
                P.append((x,y))
                x,y=0,y+1
            else: x+=1
        if len(P)==n: L.append(frozenset(P))
        if not P: break
        a,y=P.pop()
        x=a+1
    return L
```

# Affichage de la liste des solutions:

```python
def afficher_solutions(n):
    s=placer_tous_reines(n)
    i=1
    print("nomber des solutions :" ,len(s))
    for x in s:
        print("solution numero ",i,":",list(x))
        i=i+1
```

# Affichag des solutions :

```python
import numpy as np
def afficher_solutions2(n):
    s=placer_tous_reines(n) 
    print("nomber des solutions trouvees :",len(s))
    for j in range(len(s)):
        m=np.array([n,n])
        m=np.ones(m)
        m=m.astype(int)
        m=m.astype(str)
        for x in s:
            k =list(x)
        for i in range(len(k)):
            m[k[i][0]][k[i][1]]='Y'
        print("solution numero ",j+1,":\n",m)
        
```
